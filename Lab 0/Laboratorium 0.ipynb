#### 1. Import Library

```python
from skimage import color, io
from skimage.transform import rescale
import matplotlib.pyplot as plt
import os
import sys
```

#### 2. Membaca Citra

```python
i1 = io.imread('Lab0-images/Lab0-images/fasilkom.png')
```

#### 3. Menampilkan Citra 

```python
plt.imshow(i1, cmap='gray', vmin=0, vmax=255)
plt.show()
```

#### 4. Mengecek Dimensi Citra 

```python
dimension = i1.shape
dimension
```

#### 5. Menyimpan Gambar

```python
os.system('rm -f fasilkombaru.png')
io.imsave(fname='Lab0-images/Lab0-images/fasilkombaru.png', arr=i1)
```

#### 6. Visualisasi Citra 

```python
plt.imshow(X=i1, cmap='gray')
plt.title('Gambar 1')
plt.axis([0, 700, 400, 0])
plt.show()
```

#### 7. Resolusi 

```python
i3 = rescale(i1, 0.5, anti_aliasing=True)
i4 = rescale(i1, 0.1, anti_aliasing=True)
plt.imshow(X=i3, cmap='gray')
plt.show()
plt.imshow(X=i4, cmap='gray')
plt.show()
```

#### 8. Konversi Citra

```python
from skimage import data
```

```python
rgb_image = data.chelsea()
fig, ((img0, img1, img2), (img4, img5, img6)) = plt.subplots(nrows=2, ncols=3, figsize=(8,2))
img0.imshow(rgb_image)
img0.set_title('Kucing Unyu RGB Image')
img0.axis('off')

hsv_image = color.rgb2hsv(rgb_image)
img1.imshow(hsv_image)
img1.set_title('Kucing Unyu HSV Image')
img1.axis('off')

gray_image = color.rgb2gray(rgb_image)
img2.imshow(gray_image)
img2.set_title('Kucing Unyu Grayscale Image')
img2.axis('off')
fig.tight_layout()
```
